split separa cadenas de texto en función del paramaetro que le demos, en este caso no se le dio ninguno y asi por defecto separa los dos elementos

```python
inmuebles.Valor.str.split()
```

str se aplica a todda la comuna valor para convertirlo a split y separar los elementos a toda la columna


```python
inmuebles.Valor.str.split(expand=True)
```

EXPAND=TRUE lo que hace es arrojar un data frend en dos columnas.
creamos un  nuevo atributo para signo $ llamado moneda y precio para las cantidades esto crea dos nuevas columnas en la data frend de la siguiente manera:

```python
valor = inmuebles.Valor.str.split(expand=True)
inmuebles['Moneda'] = valor [0]
inmuebles['Precio'] = valor [1]
inmuebles.sample(3)
```

```python
inmuebles.info()
```

ya en la información de la data se detalla las dos nuevas columnas pero aun siguen siendo objeto la diea es cambiar el precio a numerico.Antes hay que elimnar el punto de miles  

```python
inmuebles['Precio'] = inmuebles ['Precio'].str.replace('.','',regex=True)
```

```python
inmuebles[['Precio', 'Barrio']]
```

para transformar los datos a numericos se debe usar float

---



```python
inmuebles['Precio_Millon'] = inmuebles.Precio.astype('float')/1000000
inmuebles.info()
```

ya se puede ver una nueva columna numerica float osea punto flotabnt que contiene numeros decimales o muchos mas grandes ,la columna es precio_millon

```python
inmuebles['Precio_Millon'].plot.hist(bins=50)
```

```python
import matplotlib.pyplot as plt
import seaborn as  sns

plt.figure(figsize=(10,6))
grafica = sns.histplot(data=inmuebles, x='Precio_Millon', kde=True, hue='Tipo')
grafica.set_title('Distribución de valor de inmuebles en Bogotá')
plt.xlim((50,1000))
plt.savefig('/content/drive/MyDrive/Carpeta sin título/valor_inmuebles.png',format='png')
plt.show()
```

figsize es el lienzo sobre el cual vamos a trabajar
xlim la concentración de eventos .
curva de densidad se colca en el histograma como kde=true

```python
inmuebles_filtrados = inmuebles.query("Tipo in ['Casa', 'Apartamento', 'Finca']")
plt.figure(figsize=(10,6))
grafica = sns.histplot(data=inmuebles_filtrados, x='Precio_Millon', kde=True, hue='Tipo' )
grafica.set_title('Distribución de valor de inmuebles en Bogotá')
plt.xlim((50,1000))
plt.ylim((0,100))
plt.xlabel('Rangos de precios por millón', fontsize = 14, color='red')
plt.ylabel('Inmuebles por tipos', fontsize = 14, color= 'blue')
plt.savefig('/content/drive/MyDrive/Carpeta sin título/valor_inmuebles_portipo.png',format='png')
plt.show()
```
